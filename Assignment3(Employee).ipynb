```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
emp = pd.read_csv("employee.csv")
```

```python
emp
```

```python
dep_salary = emp.groupby("Department")["Salary"].sum().sort_values(ascending=False)
plt.figure();dep_salary.plot(kind='bar')
plt.title("Salary Of Each Department"); plt.xlabel("Department"); plt.ylabel("Salary")
```

```python
plt.hist(emp['Salary'].dropna(), bins=8, edgecolor='black')
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Number of Employees")
plt.show()
```

```python
avg_salary_des = emp.groupby('Designation')['Salary'].mean()

avg_salary_des.plot(kind='bar')
plt.title("Average Salary by Designation")
plt.ylabel("Average Salary")
```

```python
dept_count = emp['Department'].value_counts()

dept_count.plot(kind='bar')
plt.title("Employee Count by Department")
plt.ylabel("Number of Employees")
plt.show()
```

```python
designation = emp['Designation'].value_counts()

designation.plot(kind='bar')
plt.title("Employee Count by Designation")
plt.ylabel("Number of Employees")
plt.show()
```

```python
avg_age_designation = emp.groupby('Designation')['Age'].mean()
plt.figure ; avg_age_designation.plot(kind="bar") ; plt.title("Average Age PER Designation")
plt.ylabel("Age")
```

```python
avg_salary_dep = emp.groupby('Department')['Salary'].mean()
plt.figure ; avg_salary_dep.plot(kind="bar") ; plt.title("Average Salary PER Department")
plt.ylabel("Salary")
```

```python
avg_salary_desig = emp.groupby('Designation')['Salary'].mean()
plt.figure ; avg_salary_desig.plot(kind="bar") ; plt.title("Average Salary PER Designation")
plt.ylabel("Salary")
```
