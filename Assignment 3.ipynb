### Type Casting

```python
# string --> int
# int --> String
# int --> float
# float --> int
# float --> str
# str --> float
```

```python
# string --> int
st = input("Enter A Number: ")
integer = int(st)
print(f"Before {type(st)} == '{st}' --> After {type(integer)} == {integer}")
```

```python
# int --> String
integer = int(input("Enter A Number: "))
string = str(st)
print(f"Before {type(integer)} == {integer} --> After {type(string)} == '{string}'")
```

```python
# int --> float
integer = int(input("Enter A Number: "))
floa = float(integer)
print(f"Before {type(integer)} == {integer} --> After {type(floa)} == {floa'}")
```

```python
# float --> int
floa = float(input("Enter A Decimal Number: "))
integer = int(floa)
print(f"Before {type(floa)} == {floa} --> After {type(integer)} == '{integer}'")
```

### mathematical formulas

```python

num_1 = int(input("Enter Number 1: "))
form = input("Enter What Will Be Done (+,-,x,/.Reminder(//),Power(**): ")
num_2 = int(input("Enter Number 2: "))
if form == "+":
    result = num_1 + num_2
    print(result)
elif form == "-":
    result = num_1 - num_2
    print(result)
elif form == "x" or "*":
    result = num_1 * num_2
    print(result)
elif form == "/":
    result = num_1 / num_2
    print(result)
elif form == "**":
    result = num_1 ** num_2
    print(result)
elif form == "//":
    result = num_1 // num_2
    print(result)
else:
    print("Enter a correct formula")
```
