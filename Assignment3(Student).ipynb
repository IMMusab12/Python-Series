```python
import pandas as pd
import matplotlib.pyplot as plt
std = pd.read_excel("student.xlsx")
```

```python
grade_counts = std['Grade'].value_counts()

plt.figure(figsize=(6,4))
plt.bar(grade_counts.index, grade_counts.values, color=['green','blue','orange'])
plt.title("Grade Distribution")
plt.xlabel("Grade")
plt.ylabel("Number of Students")
plt.show()
```
